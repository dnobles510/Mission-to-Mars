```python
# Import Splinter, BeautifulSoup, Matplotlib, and Pandas
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import matplotlib.pyplot as plt

# set up Splinter
executable_path = {'executable_path' : ChromeDriverManager().install()}
browser = Browser ('chrome', **executable_path, headless=False)

# Visit the Mars Facts site
url = 'https://data-class-mars-challenge.s3.amazonaws.com/Mars/index.html'
browser.visit(url)

html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
# Scrape table data
rows = html_soup.find_all('table', class_='table')
print(rows)
```

```python
mars_temperatures = []
for row_data in rows:
    row_heading = row_data.find_all('td')
    row_data = [col.text for col in row_heading]
    mars_temperatures.append(row_data)
print(mars_temperatures)
```

```python
# Create DataFrame
df = pd.read_html('https://data-class-mars-challenge.s3.amazonaws.com/Mars/index.html')
df
```

```python
mars_temperature_df = df[0]
mars_temperature_df
```

```python
browser.quit()
```

```python
mars_temperature_df.dtypes
```

```python
from datetime import datetime

mars_temperature_df['terrestrial_date'] = pd.to_datetime(mars_temperature_df['terrestrial_date'])
mars_temperature_df.info()
```

```python
# How many months exist on Mars?
plt.figure(figsize = (20, 5))
plt.plot(mars_temperature_df["terrestrial_date"], mars_temperature_df["min_temp"])
#plt.figure(figsize = (40, 30))
plt.show()
```

There are roughly 22-23 months that exist in Mars.

```python
# How many Martian days worth of data exists in the scraped dataset?
mars_temperature_df["sol"].nunique()
```

```python
# what are the coldest and warmest months on Mars (at the location on curiosity)
mars_month_temp = mars_temperature_df.groupby("month")['min_temp'].mean()
mars_month_temp.sort_values().plot(kind = "bar")
plt.show()
```

The 8th month is the warmest and the 3rd month is the coldest.

```python
# Which months have the lowest and highest atmospheric pressure on mars?
mars_month_pressure = mars_temperature_df.groupby("month")['pressure'].mean()
mars_month_pressure.plot(kind="bar")
plt.show()
```

The lowest atmospheric pressure month is month 6 and the highest atmospheric pressure is month 9.

```python
# About how many terrestrial (Earth) days exist in a Martian year?
plt.figure(figsize = (20, 5))
plt.plot(mars_temperature_df["sol"], mars_temperature_df["min_temp"])
plt.show()
```

There are approximately 800 earth days in a Martian year.
